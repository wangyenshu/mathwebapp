### 题目
观察1/(1+25*x^2)在[-1,1]的龙格现象，找出另一个具有龙格现象的函数
### 代码
sagemath 代码如下:

```python
def lag(x,f):
    n=len(x)
    M=matrix(n,lambda i,j: x[i]^j)
    sol=M.solve_right(f)
    return sol
R = PolynomialRing(RR, 'x')
p1=plot(1/(1+25*x^2),(x,-1,1),title='A Plot of 1/(1+25*x^2)')
p2=plot(1/(1+5^6*x^6),title='A Plot of 1/(1+5^6*x^6)')
n=int(input("Please enter a number for n: "))
x1=vector(srange(-1, 1 + 1/n, 1/n))
f1=vector(map(lambda x: 1/(1+25*x^2),x1))
lagpoly1=R(list(lag(x1,f1)))
plag1=plot(lagpoly1, xmin=-1,xmax=1,title='A Plot of Interplotation of 1/(1+25*x^2)')
f2=vector(map(lambda x: 1/(1+5^6*x^6),x1))
lagpoly2=R(list(lag(x1,f2)))
plag2=plot(lagpoly2, xmin=-1,xmax=1,title='A Plot of Interplotation of 1/(1+5^6*x^6)')

#p1.show()
#p2.show()
#p3.show()
#p4.show()
#use sage -n jupyter to run jupyter
```
